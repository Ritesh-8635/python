```python
import matplotlib.pyplot as plt

x=[45,78,51,21,26,35]
y=["part1","part2","part3","part4","part5","part6"]
color=["red","blue","green","orange","yellow"]
plt.bar(y,x,color=color,edgecolor="black")
plt.xlabel("parts of me",fontsize=20)
plt.ylabel("popularity",fontsize=20)
plt.title("popularity of movies",fontsize=20)

plt.show()
```
